**Import necessary libraries**

````python
import torch
````

# **Indexing**

````python
x = torch.rand(7).reshape(7,)
print(x)

# Indexing
y = x[1]
print(y)
````

````python
x = torch.rand((5,6))
print(x)
print("")
# Indexing
y = x[1,4]
print(y)
````

````python
x = torch.rand((5,3,4))
print(x)
print("")
y = x[1,1,3]
print(y)
````

**Indexing is a view, not a copy**

````python
y += 1
print(y)

print(x)
````

# **Slicing**

````python
x = torch.rand(12).reshape(12,)
print(x)
print()
# Indexing
y = x[1:10]
print(y)
````

````python
x = torch.arange(240).reshape(2,3,4,10)
print(x)

# Slicing
y = x[0, :, 3, 2:-1:2]
print(y)
````

**Slicing is a view, not a copy**

````python
y *= -100
print(y)

print(x)
````

# **Fancy indexing: Boolean**

````python
x = torch.arange(24).reshape(2,3,4)
print(x)

criteron = (x > 10) & (x % 2 == 1)
print(criteron)

# Boolean fancy indexing
y = x[criteron]
print(y)
print(y.shape)
````

**Boolean fancy indexing is a copy, not a view**

````python
y *= -100
print(y)

print(x)
````

# **Fancy indexing: Coordinates**

Example: `x` has dimension 4. We choose the following coordinates `x[i,j,k,l]`:

* Indexing on `(i,k) = (0,3), (1,3), (0,1), (0,1), (1,2), (0,2)`

* Slicing on `j: [:]`

* Slicing on `l: [1:-1:2]`

````python
x = torch.arange(120).reshape(2,3,4,5)
print(x)

# Coordinates fancy indexing
y = x[[0, 1, 0, 0, 1, 0], :, [3, 3, 1, 1, 2, 2], 1:-1:2]
print(y)
print(y.shape)
````

**Coordinates fancy indexing is a copy, not a view**

````python
y *= -100
print(y)

print(x)
````

# **Exercise**

Consider training a BERT model in large language models (LLM).

* Let `output_tokens` be with the shape `(num_samples,length)` and value taken from `0, 1, ..., vocab_size -1`. So `output_tokens[n,l]` is the ground-truth output from position `l` in sample `n`.

* Let `logits` be with the shape `(num_samples,length,vocab_size)`. So `logits[n,l,k]` is the predict logit of token `k` on position `l` in sample `n`.

* Let `attention_mask` be a boolean tensor with shape `(num_samples,length)`. So `attention_mask[n,l]` is `True` if position `l` in sample `n` needs to be attended (that is, we need to learn from this position). Otherwise, if `attention_mask[n,l]` is `False`, then we skip learning from position `l` in sample `n`.

Write code to compute the cross entropy loss.



````python
# Hyperparameters
num_samples = 8
length = 10
vocab_size = 4

# Instantiate ground-truth output
output_tokens = torch.randint(0, vocab_size, (num_samples, length))
print(output_tokens.shape)

# Instantiate predicted logits
logits = torch.randn(num_samples, length, vocab_size)
print(logits.shape)

# Instantiate attention_mask
attention_mask = torch.randn(num_samples, length) > 0.3
print(attention_mask.shape)

# Compute predicted log probability for each token on each position in each sample
log_prob = logits - torch.log(torch.sum(torch.exp(logits), dim=2, keepdim=True))
print(log_prob.shape)

# Compute log_prob for the ground-truth output token on each position in each sample
log_prob_true = log_prob[torch.arange(num_samples).unsqueeze(-1), torch.arange(length), output_tokens]
print(log_prob_true.shape)

# Compute the cross entropy loss on samples and positions with attention
loss = torch.mean(- log_prob_true[attention_mask])
print(loss)
````
